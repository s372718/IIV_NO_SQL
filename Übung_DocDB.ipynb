## 1. Document Databases and Object mappers

### Document DB

Although they were created many years ago, databases are now found all over the internet and are the basis of every information system.
Websites, apps, and Enterprise IS use databases to collect, store, and retrieve data.


Relational databases, which arrange data in tables with rows, have predominated the database landscape for a long time.
The relational model imposes a fixed structure, although various other database formats have arisen in recent years to overcome this constraint.

While relational databases usually utilize Structured Query Language, often known as SQL, to manage and query data, these new database types are collectively referred to as NoSQL databases.
NoSQL databases provide excellent scalability and flexibility in terms of data organization. NoSQL databases can handle massive amounts of data and quick, agile development because of their qualities.


Document databases store data as documents, breaking free from the idea that databases are made up of rows and columns, as is the case in a table inside a relational database.
Consider a document as a self-contained data entry that includes all of the information necessary to grasp its meaning, much like documents used in the real world.
